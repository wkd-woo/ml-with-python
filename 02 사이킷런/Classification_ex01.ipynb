```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
```

```python
# 붓꽃 데이터 세트를 로딩합니다.
iris = load_iris()
```

```python
# iris.data는 Iris 데이터 세트에서 피처(feature)만으로 된 데이터를 numpy로 가지고 있습니다.
iris_data = iris.data
```

```python
# iris.target은 붓꽃 데이터 세트에서 레이블(결정 값) 데이터를 numpy로 가지고 있습니다.
iris_label = iris.target
print('iris target값:', iris_label)
print('iris target명:', iris.target_names)
```

```python
# 붓꽃 데이터 세트를 자세히 보기 위해 DataFrame으로 변환합니다.
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
```

```python
iris_df['label'] = iris.target
```

## 학습용 데이터와 테스트용 데이터 분리

`학습용 데이터`와 `테스트용 데이터`는 반드시 분리해야 한다.

학습 데이터로 학습된 모델이 *얼마나 뛰어난 성능을 가지는지 평가* 하려면 **테스트 데이터 세트가 필요**하기 때문.

`sklearn`의 `train_test_split()` API를 사용.


```python
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label,
                                                    test_size=0.2, random_state=11)
```

```python
# DecisionTreeClassifier 객체 생성
dt_clf = DecisionTreeClassifier(random_state=8)
```

```python
# 학습 수행
dt_clf.fit(X_train, y_train)
```

```python
# 학습이 완료된 DecisionTreeClassifier 객체에서 테스트 데이터 세트로 예측 수행.
pred = dt_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('예측 정확도: {0:.4f}'.format(accuracy_score(y_test, pred)))
```

### 붓꽃 데이터 세트로 분류를 예측한 프로세스 정리

1. 데이터 세트 분리: 데잍를 학습 데이터와 테스트 데이터로 분리한다.
2. 모델 학습: 학습 데이터를 기반으로 ML 알고리즘을 적용해 모델을 학습시킨다.
3. 예측 수행: 학습된 ML 모델을 이용해 테스트 데이터의 분류(즉, 붓꽃 종류)를 예측한다.
4. 평가: 이렇게 예측된 결괏값과 테스트 데이터의 실제 결괏값을 비교해 ML 모델 성능을 평가한다.
